```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Titanic-Dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Gender vs Survival")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Passenger Class vs Survival")
plt.show()
```

```python
sns.histplot(df['Age'].dropna(), bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Fare Outliers")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

1.

Female passengers had significantly higher survival rates compared to males.

2.

Passengers traveling in 1st class had better survival chances than lower classes.

3.

Most passengers belonged to the 20–40 age group.

4.

Higher ticket fares may indicate wealthier passengers who had better access to safety measures.

5.

A large number of missing age values suggests incomplete passenger records.

6.

Socio-economic status strongly influenced survival probability.
