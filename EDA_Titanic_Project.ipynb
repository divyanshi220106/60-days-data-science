```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class")
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Fare Outliers")
plt.show()
```

**Insights**

1. Female passengers survived more than male passengers.
2. Most passengers were between 20–40 years old.
3. First-class passengers had better survival rates.
4. Fare column contains several outliers.
5. Third-class passengers were highest in number.
